A olayının B koşullu olasılığı, P(A  | B) ile gösterilir. Koşullu olasılık formülü olarak yukarıdaki eşitlikleri kullanabiliriz.

**ÖRNEK:** Bir okulun öğrencilerinden %45’i fizik, %35’i matematik derslerinden ve %25’i hem fizik hem de matematik derslerinden başarısızdır. Rasgele seçilen bir öğrencinin,

a) Fizikten başarısız olduğu biliniyorsa, matematikten de başarısız olma olasılığını bulunuz

b)Matematikten başarısız olduğu biliniyor ise, fizikten de başarısız olma olasılığını bulunuz

**ÇÖZÜM:** 

M = Matematikten başarısız olanların sayısı
F = Fizikten başarısız olanların sayısı

**P(M) = 0,35**

**P(F) = 0,45**

**P(M ∩ F) = 0,25**


image.png



```python
def FizikOlasilik():
  pM = 0.35
  pF = 0.45
  pMnF = 0.25
  pMF = pMnF /  pF
 
  return pMF

FizikOlasilik()
```

```python
def MatematikOlasilik():
  pM = 0.35
  pF = 0.45
  pMnF = 0.25
  pFM = pMnF /  pM
 
  return pFM

MatematikOlasilik()
```

# **BAYES TEOREMİ**

image.png



**P(A|B) =** B olayı gerçekleştiğinde A olayının gerçekleşme olasılığı

**P(A) = A** olayının gerçekleşme olasılığı

**P(B|A) = A** olayı gerçekleştiğinde B olayının gerçekleşme olasılığı

**P(B) = B** olayının gerçekleşme olasılığı

**ÖRNEK:** Bir atölyede imal edilen ürünlerin istenilen normlara uyan iyi bir ürün olarak kabul edilme olasılığı, 0,96’dır. Yapılan bir teknik kontrol sistemine göre, iyi bir ürün için, bunun iyi olduğunu ortaya çıkarma olasılığı 0,98 dir. Kötü olan bir ürünün iyi olduğu yolunda karar verme olasılığı 0,05’dir. İki kez yapılan çekime göre bir ürünün iyi olduğu sonucuna varıldığına göre, bu ürünün gerçekten iyi olma olasılığı nedir?

**ÇÖZÜM:**

**B1 :** ürünü iyi kabul etme 

**B2 :** ürünü kötü kabul etme 

**A:** ürünün iyi olması kararı 

**P(B1) =** 0,96 

**P(B2) =** 0,04 

**P(A/B1) =** 0,98 

**P(A/B2) =** 0,05

Bayes formülüne göre;

image.png

```python
def BayesTeorisi():
  pB1 = 0.96
  pAB = 0.98
  pB2 = 0.04
  pB_ = 0.05
  pS = pB1 * pAB + pB2 * pB_ 
  return pB1 * pAB / pS

print(BayesTeorisi())
```
